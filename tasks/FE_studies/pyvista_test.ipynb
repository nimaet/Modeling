```python
import numpy as np
import pyvista as pv

# Data
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

# Create a PolyData line
line = pv.PolyData(np.column_stack((x, y, np.zeros_like(x))))
line["y"] = y

# Plot
plotter = pv.Plotter()
plotter.add_mesh(
	line,
	scalars="y",
	line_width=3
)
plotter.show()
```

```python
plotter = pv.Plotter()
plotter.add_mesh(
	line,
	color="black",
	line_width=3
)

plotter.view_xy()
plotter.enable_parallel_projection()
plotter.show_axes()
plotter.show()
```

```python
import numpy as np
import pyvista as pv

x = np.linspace(0, 2*np.pi, 400)
points = np.column_stack((x, np.sin(x), np.zeros_like(x)))
line = pv.PolyData(points)

plotter = pv.Plotter()
plotter.add_mesh(line, color="black", line_width=3)

plotter.view_xy()
plotter.enable_parallel_projection()
plotter.show_axes()

# --- start recording ---
plotter.open_movie("sine_wave.mp4")


plotter.show(auto_close=False)

for t in np.linspace(0, 4*np.pi, 200):
	points[:, 1] = np.sin(x - t)
	line.points = points
	plotter.write_frame()    # save frame
	plotter.render()

plotter.close()
```
