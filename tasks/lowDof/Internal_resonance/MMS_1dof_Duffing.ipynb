```python
import sympy as sp

# symbols
t, eps, delta = sp.symbols('t eps delta')
T0, T1 = sp.symbols('T0 T1')

# unknown functions
x0 = sp.Function('x0')(T0, T1)
x1 = sp.Function('x1')(T0, T1)

# expansion
x = x0 + eps*x1

# derivative operators
D0 = lambda f: sp.diff(f, T0)
D1 = lambda f: sp.diff(f, T1)

# total derivatives
dx_dt = D0(x) + eps*D1(x)
d2x_dt2 = D0(D0(x)) + 2*eps*D0(D1(x))

# equation
eq = d2x_dt2 + x + eps*x**3

# expand
eq_series = sp.expand(eq)

# extract epsilon orders directly (robust with mixed derivatives)
order0 = eq_series.coeff(eps, 0)
order1 = eq_series.coeff(eps, 1)

print("O(1) equation:")
print(order0)

print("\nO(eps) equation:")
print(order1)
```

```python
A = sp.Function('A')(T1)
x0 = A*sp.exp(sp.I*T0) + sp.conjugate(A)*sp.exp(-sp.I*T0)

forcing = -2*sp.diff(x0, T0, T1) - x0**3

# expand
forcing = sp.expand(forcing)

# find coefficient of exp(iT0)
coef = sp.simplify(sp.expand(forcing).coeff(sp.exp(sp.I*T0)))
print(coef)
```
